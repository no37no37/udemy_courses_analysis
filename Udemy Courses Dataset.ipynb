```python
# data.dtypes
# data['published_timestamp'] = pd.to_datetime(data.published_timestamp)
# data.dtypes
# data['Year'] = data['published_timestamp'].dt.year
# data[data.year == 2015]
```

```python
data.head(2)
```

```python
data.dtypes

# we need to convert "published_timestamp" to datetime
```

```python
data['published_timestamp'] = pd.to_datetime(data.published_timestamp)

# pd.to_datetime - this function converts value of a column to datetime
```

```python
data.dtypes
```

```python
# now we can separate the year from this datetime column

# data['Year'] - syntax of creating a new column

data['Year'] = data['published_timestamp'].dt.year
```

```python
data[data.Year == 2015]
```
